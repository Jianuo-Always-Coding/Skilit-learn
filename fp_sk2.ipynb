```python
import pandas as pd
import numpy as np
```

```python
beike = pd.read_csv('./data.csv', encoding='utf-8-sig')
```

```python
beike['房屋户型']
```

```python
# 去除房屋户型为“车位”的那一部分
beike_type=beike[beike['房屋户型']!='车位']
beike_type
```

```python
beike_type['房屋户型']
```

```python
beike_type['houseType']=beike_type['房屋户型'].apply(lambda x: x.replace('bed_','室,').replace('liv_','厅,').replace('kit_','厨,'))
beike_type['houseType']
```

```python
beike_type
```

```python
# 关于tfidf的reference：https://www.jianshu.com/p/c7e2771eccaa
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(beike_type['roomType1'])
tfidf_matrix.shape
```

```python
print(tfidf_matrix)
```

```python
tfidf.vocabulary_
```

```python
tfidf_weight = tfidf_matrix.toarray()
tfidf_weight
```

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=50,random_state=10)
s = clf.fit(tfidf_weight)
```

```python
beike_type['labels']=clf.labels_
```

```python
beike_type
```

```python
yuexiu = beike_type[beike_type['城市区']=='越秀'].loc[:,('小区坐标','元/平','成交时间','labels')].reset_index(drop=True)
for i in range(yuexiu.shape[0]):
    yuexiu['元/平'][i]=yuexiu['元/平'][i][:-4]
yuexiu
```

```python
yuexiu['lon']=yuexiu['小区坐标'].apply(lambda x: x.split(',')[0])
yuexiu['lat']=yuexiu['小区坐标'].apply(lambda x: x.split(',')[1])
yuexiu.drop(columns='小区坐标',inplace=True)
yuexiu
```

```python
yuexiu['成交年份']=yuexiu['成交时间'].apply(lambda x: x.split('.')[0])
yuexiu.drop(columns='成交时间',inplace=True)
yuexiu
```

```python
# 由高德地图（https://lbs.amap.com/tools/picker）获取广州市政府所在地坐标为113.264385,23.129112
# 取经纬度差值为0.01处（约1km）
center_lon = 113.264385
center_lat = 23.129112
cost = []
label = []
year = []
for i in range(yuexiu.shape[0]):
    lon = float(yuexiu['lon'][i])
    lat = float(yuexiu['lat'][i])
    if abs(lon-center_lon) < 0.01 and abs(lat-center_lat) < 0.01:
        cost.append(int(yuexiu['元/平'][i]))
        label.append(yuexiu['labels'][i])
        year.append(yuexiu['成交年份'][i])
```

```python
data = {'yuan/m2':cost,'year':year,'label':label}
centerHouse_info = pd.DataFrame(data)
centerHouse_info
```

```python
cal_mean = centerHouse_info.groupby(['year','label'], as_index=False)['yuan/m2'].mean()
cal_mean
```

```python
label_list = list(set(label))
for i in range(len(label_list)):
    if cal_mean[cal_mean['label']==label_list[i]].shape[0] == 6:
        print(cal_mean[cal_mean['label']==label_list[i]])
```
